**1.Print customer Id, customer name and average account_balance maintained by each customer for all of his/her accounts in the bank.(8 Rows)**

```python
query = """
SELECT
    a.customer_id,
    a.customer_name,
    AVG(b.Balance_amount) AS avg_balance
FROM BANK_CUSTOMER a
LEFT JOIN Bank_Account_Details b
    ON a.customer_id = b.Customer_id
GROUP BY a.customer_id, a.customer_name
LIMIT 8;
"""

cursor.execute(query)

columns = [desc[0] for desc in cursor.description]
result = cursor.fetchall()

print("{:<12} {:<20} {:<15}".format(*columns))
print("-" * 50)

for row in result:
    print("{:<12} {:<20} {:<15.2f}".format(*row))
```

**# 3) Print account_number and balance_amount , transaction_amount,Transaction_Date from Bank_Account_Details and bank_account_transaction for all the transactions occurred during march,2020 and april, 2020(12 Rows)**

```python
import pandas as pd
from tabulate import tabulate

query = """
SELECT
    a.Account_Number,
    a.Balance_amount,
    t.Transaction_amount,
    t.Transaction_Date
FROM Bank_Account_Details a
JOIN BANK_ACCOUNT_TRANSACTION t
      ON a.Account_Number = t.Account_Number
WHERE strftime('%m', t.Transaction_Date) IN ('03', '04')
  AND strftime('%Y', t.Transaction_Date) = '2020'
LIMIT 12;
"""

# Read SQL into pandas DataFrame
df = pd.read_sql_query(query, conn)

# Print table
print(tabulate(df, headers='keys', tablefmt='grid'))
```

**4) Print all of the customer id, account number,  balance_amount, transaction_amount , Transaction_Date  from bank_customer, Bank_Account_Details and bank_account_transaction tables where excluding all of their transactions in march, 2020  month (22 Rows)**

```python
import pandas as pd
from tabulate import tabulate

query = """
SELECT
    c.customer_id,
    a.Account_Number,
    a.Balance_amount,
    t.Transaction_amount,
    t.Transaction_Date
FROM bank_customer c
JOIN Bank_Account_Details a
      ON c.customer_id = a.Customer_id
LEFT JOIN bank_account_transaction t
      ON a.Account_Number = t.Account_Number
     AND NOT (
            strftime('%m', t.Transaction_Date) = '03'
        AND strftime('%Y', t.Transaction_Date) = '2020'
     )
ORDER BY c.customer_id, a.Account_Number
LIMIT 22;
"""

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt='psql'))
```

**# 7) Print all Customer_id, Account_Number, Account_type, and display deducted balance_amount by  subtracting only negative transaction_amounts for Relationship_type ="P" ( P - means  Primary , S - means Secondary ) .(27 Rows)**

```python
import pandas as pd
from tabulate import tabulate

query = """
SELECT
    a.Customer_id,
    a.Account_Number,
    a.Account_type,
    a.Balance_amount
        - COALESCE(SUM(
                CASE
                    WHEN t.Transaction_amount < 0 THEN ABS(t.Transaction_amount)
                    ELSE 0
                END
            ), 0) AS deducted_balance
FROM Bank_Account_Details a
LEFT JOIN bank_account_transaction t
       ON a.Account_Number = t.Account_Number
WHERE a.Relationship_type = 'P'
GROUP BY
    a.Customer_id,
    a.Account_Number,
    a.Account_type,
    a.Balance_amount
ORDER BY a.Customer_id, a.Account_Number;
"""

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt='psql'))
```

**# 10) Display all saving account holders have â€œAdd-on Credit Cards" and â€œCredit cards" (3 Rows)**

```python
import pandas as pd
from tabulate import tabulate

query = """
SELECT DISTINCT
    s.Customer_id,
    s.Account_Number AS Savings_Account,
    cc.Account_Number AS Credit_Card,
    acc.Account_Number AS Addon_Credit_Card
FROM Bank_Account_Details s
JOIN Bank_Account_Details cc
     ON s.Customer_id = cc.Customer_id
    AND cc.Account_type = 'Credit Card'
JOIN Bank_Account_Details acc
     ON s.Customer_id = acc.Customer_id
    AND acc.Account_type = 'Add-on Credit Card'
WHERE s.Account_type = 'SAVINGS';
"""

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers="keys", tablefmt="psql"))
```

```python
import pandas as pd
from tabulate import tabulate

query = """
SELECT
    s.Customer_id,
    s.Account_Number AS savings_account,
    cc.Account_Number AS credit_card_account,
    SUM(t.Transaction_amount) AS total_credit_card_transactions
FROM Bank_Account_Details s
JOIN Bank_Account_Relationship_Details r
      ON s.Account_Number = r.Linking_Account_Number
JOIN Bank_Account_Details cc
      ON r.Account_Number = cc.Account_Number
JOIN bank_account_transaction t
      ON cc.Account_Number = t.Account_Number
WHERE s.Account_type = 'SAVINGS'
  AND cc.Account_type = 'Credit Card'
GROUP BY s.Customer_id, s.Account_Number, cc.Account_Number;
"""

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt='psql'))
```
